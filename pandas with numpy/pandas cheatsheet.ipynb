# Merring joinign and concatination

```python
#Merger two dataframe
```

```python
salaries = pd.DataFrame({
    "employes_id":[1,2,3,6,7],
    "salary" : [6000,7000,8000,9000,60000],
    "bonus" : [5000,10000,7000,8000,12000]
})

employees=pd.DataFrame({
    'employes_id' : [1,2,3,4,5],
    "name" :  ["john","Anna","petter","linda","pooiinn"],
    "department": ["HR","IT","FINANCE","IT","HR"]
})
```

```python
salaries
```

```python
employees
```

```python
df4=pd.merge(employees,salaries,on = 'employes_id',how='outer')
```

```python
df4
```

```python
df4.fillna(0,inplace=True)
df4
```

```python
df4.dropna(thresh=3)
```

```python
df5=df4["name"].fillna("camela"),[df4["department"].fillna("sw")]
df5
```
