MISSING DATA

```python
import numpy as np 
import pandas as pd
```

Finding Missing Data

```python
data={
    "A":[1,2,np.nan,4,5],
    "B":[1,2,3,4,5],
    "C":[1,2,3,np.nan,np.nan],
    "d":[1,np.nan,np.nan,np.nan,5]
    
}
cf=pd.DataFrame(data)
```

```python
cf.isna() # is use to find the missing data which return the missing values in True.
```

```python
cf.isna().sum()
```

```python
cf.isna().any()  # any is use to find the null values in pandas
```

Removing the Null values in the Missing Data

```python
cf
```

```python
cf.dropna(thresh=3)  # thresh is used to remove all the null  values 
```

Fill the missing values


```python
cf.fillna(0)
```

```python
values = {'A' : 0 , 'B' : 100 , 'C' : 300 , 'D' : 400}
cf.fillna(value=values) #fillna is used to fill the null value  
```

```python
cf
```

```python
cf.fillna(cf.mean())
```
