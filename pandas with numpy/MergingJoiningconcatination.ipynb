```python
import numpy as np
import pandas as pd
```

Merging 2 dataframe

```python
employees = pd.DataFrame({
    "employee_id":[1,2,3,4,5],
    "name":["john","Anna","Peter","linda","bob"],
    "department":["HR","IT","Finance","IT","HR"]
})
# DATAFRAME 2 : SALARY information
salaries = pd.DataFrame({
    "employee_id":[1,2,3,6,7],
    "salary":[60000,85000,80000,70000,90000],
    "Bonus":[5000,10000,7000,12000,8000]
})
```

```python
employees
```

```python
salaries
```

```python
# there are two variations we can concatinate these two data frame
```

```python
pd.merge(employees,salaries,on="employee_id",how="inner")
```

```python
pd.merge(employees,salaries,on="employee_id",how="outer") # outer merge both the data frame and display empty valuues as NAN
```

```python
pd.merge(employees,salaries,on="employee_id",how="left") # lefts means i merge the data on the basis of employes
```

```python
pd.merge(employees,salaries,on="employee_id",how="right") 
```

CONCATINATION OF 2 DataFrame

```python
 df1=pd.DataFrame({
     "A":["A0","A1","A2"],
     "B":["B0","B1","B2"],
     "C":["C1","C2","C3"]
 })                                   # LIST {}
 df2=pd.DataFrame({
     "A":["A3","A4","A5"],
     "B":["B3","B4","B5"],
     "C":["C3","C4","C5"]
 })
```

```python
df1
```

```python
df2
```

```python
#  Concatination functions (concat)
```

```python
pd.concat([df1,df2]) # CONCATE BOTH THE DATA FRAME
```

```python
pd.concat([df1,df2],axis=1)
```

**joining 2 Dataframes

```python
df1=pd.DataFrame({
    "name":["Abhay","Akshit","charlie"]
},index = [1,2,3])
# second dataFrame 
df2=pd.DataFrame({
    "score":[85,90,75]
},index=[1,2,3])
```

```python
df1
```

```python
df2
```

```python
# i want to joint df1 +df2
```

```python
df1.join(df2,how="outer")
```

```python
df2.join(df1)
```
