# Group By

```python
data={
    "category":["A","B","A","B","A","B","A","B"],
    "store":["S1","S2","S2","S1","S1","S1","S2","S1"],
    "sale":[100, 200, 150, 300, 450, 670, 210, 300],
    "quantity":[10,15,12,18,8,20,15,25],
    "Date":pd.date_range('2026-01-01',periods=8)
}
df=pd.DataFrame(data)
```

```python
df
```

Group create an object is does not create a data frames



# group by category and calculate the sum of state

```python
cat=df.groupby("category")["sale"].sum()
```

```python
cat
```

```python
cat=df.groupby("category")["quantity"].sum()
cat
```

```python
# group by stores and calculate the sum of sales
cat=df.groupby("store")["sale"].sum()
cat
```

# group by multiple columns
# Group by category and store

```python
# group by multiple columns
# Group by category and store
cat=df.groupby(["category","store"])['sale'].sum()
cat
```
