# Module 4: Arithmetic Operations & Broadcasting
Element-wise arithmetic, broadcasting rules, and deep vs shallow copying.

```python
import numpy as np
```

## 4.1 Element-wise Arithmetic

```python
a1 = np.array([1, 2, 3, 4, 5])
a2 = np.array([2, 3, 4, 5, 6])

print('a1 + a2 =', a1 + a2)
print('a1 - a2 =', a1 - a2)
print('a1 * a2 =', a1 * a2)
print('a1 / a2 =', a1 / a2)
print('a1 % a2 =', a1 % a2)   # element-wise modulo
```

## 4.2 Broadcasting
Broadcasting lets NumPy operate on arrays of different shapes by 'stretching' the smaller one.

```python
L = [12, 21, 12, 32, 43, 65]
arr = np.array(L)

print('arr       :', arr)
print('arr + 10  :', arr + 10)    # scalar broadcast over entire array
print('arr * 2   :', arr * 2)
```

```python
# Broadcasting on a 2D array
a2d = np.arange(1, 26).reshape(5, 5)
print('Original:\n', a2d)
print('\na2d + 10:\n', a2d + 10)
```

## 4.3 Deep vs Shallow Copying
⚠️ **Key concept**: slicing returns a *view* (shallow copy). Changes affect the original.
Use `.copy()` for an independent copy (deep copy).

```python
a = np.arange(1, 21)
print('Original a:', a)

# Shallow copy / view
view_slice = a[:5]
view_slice *= 7                        # modifies `a` too!
print('After view_slice *= 7, a:', a)  # first 5 elements changed

# Deep copy
a = np.arange(1, 21)                   # reset
deep_copy = a[:5].copy()
deep_copy *= 7
print('After deep_copy *= 7,  a:', a)  # `a` is unchanged
```

```python
# Quick demonstration with a duplicate-values array
a = np.array([1, 2, 2, 2, 2, 2])
print('Array with duplicates:', a)
print('Unique values:', np.unique(a))
```
