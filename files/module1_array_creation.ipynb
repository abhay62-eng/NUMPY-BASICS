# Module 1: NumPy Array Creation
This module covers the fundamentals of creating NumPy arrays using various methods.

```python
import numpy as np
print('NumPy version:', np.__version__)
```

## 1.1 Creating Arrays from Python Lists

```python
# 1D array (vector)
arr = np.array([1, 2, 3, 4, 5])
print('1D array:', arr)
```

```python
# Mixed types: NumPy upcasts all elements to a common type (str here)
# FIX: the original code stored a plain list, not a numpy array
a_mixed = np.array([1, 2, 3, 4, 5, 'hello'])
print('Mixed array (upcasted to str):', a_mixed)
print('dtype:', a_mixed.dtype)
```

```python
# 2D array (matrix / vector of vectors)
l = [[1, 2, 3], [4, 5, 6], [7, 6, 5]]
matrix = np.array(l)
print('2D array (matrix):\n', matrix)
```

## 1.2 Array Generation Functions

```python
# np.arange: evenly spaced values in a range
print(np.arange(1, 10))        # step defaults to 1
print(np.arange(1, 11, 2))     # step = 2
```

```python
# np.zeros / np.ones
print('zeros 1D:', np.zeros(6))
print('zeros 2D:\n', np.zeros((6, 6)))
print('ones 2D:\n',  np.ones((4, 4)))
```

```python
# np.linspace: evenly spaced values between start and stop (inclusive)
a = np.linspace(1, 5, 3)
print('linspace(1,5,3):', a)

a100 = np.linspace(1, 5, 100)
print('linspace(1,5,100) first 5:', a100[:5], '...')
```

## 1.3 Random Number Generation

```python
# Uniform random [0, 1)
print('rand vector (10):',  np.random.rand(10))
print('rand matrix (3x3):\n', np.random.rand(3, 3))

# Standard normal distribution
print('randn(10):', np.random.randn(10))

# Random integers
print('randint(10,20,10):', np.random.randint(10, 20, 10))
```
