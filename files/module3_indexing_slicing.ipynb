# Module 3: Indexing & Slicing
Covers 1D/2D indexing, slicing, and boolean indexing.

```python
import numpy as np
```

## 3.1 1D Indexing & Slicing

```python
arr = np.arange(11, 21)   # [11, 12, ..., 20]
print('array          :', arr)
print('arr[6]         :', arr[6])     # single element
print('arr[:]         :', arr[:])     # full array
print('arr[3:9]       :', arr[3:9])   # slice
print('arr[3::2]      :', arr[3::2])  # every 2nd element from index 3
print('arr[1:4]       :', arr[1:4])
```

## 3.2 2D (Matrix) Indexing & Slicing

```python
# Shape: 6 rows × 5 cols, values 1‒30
arr = np.arange(1, 31).reshape(6, 5)
print('Full matrix:\n', arr)
```

```python
print('Single element arr[0,4]        :', arr[0, 4])
print('Entire row 1 arr[1]            :', arr[1])
print('Entire column 2 arr[:,2]       :', arr[:, 2])
print('\nRows 0-1, Cols 1-2 arr[0:2,1:3]:\n', arr[0:2, 1:3])
print('\nRows 2-4, Cols 3+ arr[2:5,3:] :\n', arr[2:5, 3:])
print('\nRows 3+, Cols 3+  arr[3:,3:]  :\n', arr[3:, 3:])
```

```python
# FIX: original used arr[3:6,3:6] on a (6,5) array — col index 5 is out of range.
# Safe equivalent: slice up to last available column
safe_slice = arr[3:6, 3:5]   # rows 3-5, cols 3-4
print('Safe submatrix arr[3:6,3:5]:\n', safe_slice)
```

## 3.3 Boolean Indexing

```python
arr = np.arange(1, 20)
print('array:', arr)

# Select even numbers
bool_index = arr % 2 == 0
print('boolean mask:', bool_index)
print('even numbers:', arr[bool_index])
```

```python
# Larger range: even numbers from 10 to 99
arr = np.arange(10, 100)
evens = arr[arr % 2 == 0]
print('Evens from 10–99:', evens)
```
